## Import libraries

```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
```

## Load the data

```python
# From Scikit-learn we import the load_digits data set into the digits object
digits = load_digits()
```

```python
digits.data[0]
```

```python
# Reshape with numpy to be 8 by 8 since this is how the numbers can be visualized
image = np.reshape(digits.data[10],(8,8))
image
```

```python
# Display the image of the data reshaped with a gray scale
plt.imshow(image,cmap='gray')
```

```python
# Stores the target value of the data, here you can check the expected result
digits.target[10]
```

## Split data in train and test

```python
from sklearn.model_selection import train_test_split
# Train, test split of the x and y data
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=0)
x_train.shape
```

## Entrenamiento de modelo

```python
# Creates the model
logistic_reg = LogisticRegression(max_iter=10000)
# We train the model with the data
logistic_reg.fit(x_train,y_train)
```

## Evaluation of the model

```python
# Makes a prediction using the tests
predictions = logistic_reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
# We assure its accuracy by checking a confusion_matrix, it says how much are
# false positives, wrong and predicted wrightly.
cm = confusion_matrix(y_test,predictions)
```

```python
cm
```

```python
# Creates a new figure with a size of 9 by 9
plt.figure(figsize=(9,9))
# We define a heatmap using the confusion matrix data
sns.heatmap(cm, annot=True, linewidths=.5, square=True, cmap='coolwarm')
plt.ylabel('actual label')
plt.xlabel('Predicted label')
# Other way to check the accuracy, much simpler but faster for some cases
logistic_reg.score(x_test,y_test)*100
```
